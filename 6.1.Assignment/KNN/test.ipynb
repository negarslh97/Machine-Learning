```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100

apple_width = np.random.normal(6,1,n)
apple_length = np.random.normal(6,1,n)
banana_width = np.random.normal(4,1,n)
banana_length = np.random.normal(8,1,n)
```

```python
x_width = 3
x_length = 9

y_width = 7
y_length = 5

z_width = 5
z_length = 7
```

```python
plt.scatter(apple_width,apple_length,c='red')
plt.scatter(banana_width,banana_length,c='orange')
plt.scatter(x_width,x_length,c='blue')
plt.scatter(y_width,y_length,c='green')
plt.scatter(z_width,z_length,c='black')
plt.title('Apple and Banana')
plt.xlabel('Width')
plt.ylabel('Length')
plt.legend()
plt.show()
```

## KNN

```python
width = np.concatenate((apple_width,banana_width))
length = np.concatenate((apple_length,banana_length))
```

```python
X_train = np.array([width,length]).T
X_train
```

```python
y_apples = np.zeros(n , dtype=int)
y_banana = np.ones(n , dtype=int)

y_train = np.concatenate((y_apples,y_banana))
y_train
```

```python
def euclidean_distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
new_fruit = np.array([5 , 7])
```

```python
distances = []
for x in X_train:
    d = euclidean_distance(new_fruit,x)
    distances.append(d)

distances
```

```python
k = 5
```

```python
np.sort(distances)
```

```python
nearest_neighbors = np.argsort(distances)[0:k]
nearest_neighbors
```

```python
y_train[nearest_neighbors]
```

```python
result = np.bincount(y_train[nearest_neighbors])
```

```python
if np.argmax(result) == 0 :
    print('Apple')
else:
    print('Banana')
```
