```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
import sys
import os

current_dir = os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, '..'))
sys.path.append(parent_dir)

from perceptron import Perceptron
```

```python
X, y, coef = make_regression(n_samples=500, n_features=1, n_informative=1, noise=10, coef=True ,random_state=42)
```

```python
X = np.interp(X, (X.min(), X.max()), (1, 30))
y = np.interp(y, (y.min(), y.max()), (70000, 150000))
```

```python
plt.scatter(X, y, color='#40E0D0', label='Data points')
plt.title("Employee's Salary Dataset")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary 💵")
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Perceptron(.0001, .1, 5)
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
```

```python
(ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))[1]
ax1.scatter(X_train, y_train, color="#40E0D0")
ax1.scatter(X_test, y_test, color="#c18bdb")
ax1.plot(X_train, y_pred, color="red")
ax1.legend(['Train data','Training line','Test data'])
ax1.set_xlabel("Years")
ax1.set_ylabel("Salary")
ax1.set_title("Experience And Salary")

ax2.plot(model.losses, color="#c70039")
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Loss")
ax2.set_title("Loss And Epochs With 5 Epochs")
plt.show()
```
